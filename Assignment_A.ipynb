## Question 3
Define a variable ‘Debt’ and assign a value of 30000 to it. Write a loop that loops until the condition: ‘Debt’ larger than 6000 is not any longer true. In the loop, reduce ‘Debt’ in each step by 17% - where a step represents a year. How many years does it take until the ‘Debt’ is 6000 or less?

(Note: the base for the 17% is always the previous year's remaining debt, not the initial debt)

Please provide the answer as an integer value without any decimal places e.g. 6. 

```python
debt = 30000
i = 0

while debt > 6000:
    print(f"Year {i}: {int(debt)} remaining")
    debt -= (debt * 0.17)
    i += 1

print(f"\nYear {i}: {int(debt)} remaining (less than 6000)")
```

## Question 4
Study the following simple function:
<pre>
def my_function2(N):
    for i in range(N):
        b = i**2
        print(b)
</pre>
What will be the output of the function when it is called as my_function2(N)?

1. It will print the numbers 1 to N
2. It will print the values of "i" multiplied by 2
3. It will print the square roots of the numbers 1 to N
4. It will print the square of the numbers 1 to N
5. It will print nothing
6. It will calculate the square roots of the numbers 1 to N, but not display the result
7. It will calculate the square of the numbers 1 to N, but not display the result
8. None of the above

The function above will do none of the above, so the correct answer is 8.<br>
The function loops through the values from `0` to `N - 1`, assigning the current iteration of the loop to the variable `i`. This value is squared and assigned in `b`, which is then printed.


```python
def my_function2(N):
    for i in range(N):
        b = i**2
        print(b)

my_function2(4) # 0^2, 1^2, 2^2, 3^2
```

## Question 6

<pre>

Study the structure of a dataset called "diamonds" ( Seaborn built-in data set) . Feel free to also study the definition of the variables from https://www.geeksforgeeks.org/data-science/seaborn-datasets-for-data-science/#5-diamonds-dataset 

Using some of the variables in the data set, the following figure has been produced.

Using your knowledge in Python, replicate the above figure.

```python
diamonds = sns.load_dataset("diamonds")

fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# Price vs Carat (upper left)
ax[0,0].scatter(diamonds["carat"], diamonds["price"], alpha=0.5, color='c', edgecolor='black')
ax[0,0].set_xlabel("carat")
ax[0,0].set_ylabel("price")
ax[0,0].set_title("Price vs Carat")

# Price vs Depth (upper right)
ax[0,1].scatter(diamonds["depth"], diamonds["price"], alpha=0.5, color='m', edgecolor='black')
ax[0,1].set_xlabel("depth")
ax[0,1].set_ylabel("price")
ax[0,1].set_title("Price vs Depth")

# Price vs Table (lower left)
ax[1,0].scatter(diamonds["table"], diamonds["price"], alpha=0.5, color='y', edgecolor='black')
ax[1,0].set_xlabel("table")
ax[1,0].set_ylabel("price")
ax[1,0].set_title("Price vs Table")

# Price vs X (lower right)
ax[1,1].scatter(diamonds["x"], diamonds["price"], alpha=0.5, color='orange', edgecolor='black')
ax[1,1].set_xlabel("x")
ax[1,1].set_ylabel("price")
ax[1,1].set_title("Price vs X")

plt.tight_layout()
plt.show()
```

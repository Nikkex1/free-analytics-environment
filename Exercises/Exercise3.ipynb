```python
# Imports
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Python exercises for data visualization

```python
# 1
meta = pd.read_csv("Meta_prices.csv",sep=",")

print(meta.describe())
print(meta.info())
print(meta.head())
```

```python
# i, ii)
fig1, ax1 = plt.subplots()

ax1.scatter(meta["Open"],meta["Close"],
            color="blue",marker="o")

ax1.set_xlabel("Open price")
ax1.set_ylabel("Close price")
ax1.set_title("Open vs. Close Prices of Meta Stock")

plt.show()
```

iii) <br>
There is a linear relationship (positive correlation) between the open and close prices of the Meta stock. Based on the scatterplot, no outliers are present.

```python
# 2

# i)
fig2, ax2 = plt.subplots()

ax2.plot(meta["Date"],meta["Close"])
ax2.tick_params(axis="x",labelrotation=45)

plt.show()
```

ii)<br>
The x-axis labels are incomprehensible, as there are too many of them.

```python
# iii)
meta["Date"] = pd.to_datetime(meta["Date"])

fig2, ax2 = plt.subplots()

ax2.plot(meta["Date"],meta["Close"],"g--",marker="+")
ax2.tick_params(axis="x",labelrotation=45)

ax2.set_xlabel("Date")
ax2.set_ylabel("Closing Price")
ax2.set_title("Meta Stock Closing Prices Over Time")

plt.show()
```

The use of a datetime range made the x-axis clear and understandable.

```python
# 3

# i)
flights = sns.load_dataset("flights")

print(flights.describe())
print(flights.info())
print(flights.head())
```

```python
# ii)
mask1 = flights["year"] == 1950
flights_1950 = flights[mask1]

# iii, iv)
fig3, ax3 = plt.subplots()

ax3.plot(flights_1950["month"],flights_1950["passengers"],
         "g-.",
         marker="s",markerfacecolor="r",markeredgecolor="r")

ax3.tick_params(axis="x",labelrotation=45)
ax3.set_xlabel("Month")
ax3.set_ylabel("Passengers")
ax3.set_title("Number of Passengers by Months")

plt.show()

# v)
mask2 = flights["year"] == 1960
flights_1960 = flights[mask2]

# vi)
fig4, ax4 = plt.subplots()

ax4.plot(flights_1950["month"],flights_1950["passengers"],
         color="skyblue",linestyle=":",
         marker="d",markerfacecolor="g",markeredgecolor="g",
         label="1950")
ax4.plot(flights_1960["month"],flights_1960["passengers"],
         color="skyblue",linestyle="-",
         marker="s",markerfacecolor="r",markeredgecolor="r",
         label="1960")

ax4.tick_params(axis="x",labelrotation=45)
ax4.set_xlabel("Month")
ax4.set_ylabel("Passengers")
ax4.set_title("Number of Passengers by Months")

plt.legend()
plt.show()
```

The amount of passengers per month has increased tremendously between 1950 and 1960. The lineplot also hints at correlation between passenger amount and month (more flights in summer months).

```python
# 4

# i, ii)
flights_grouped = flights.groupby("year")["passengers"].sum().reset_index()

fig5, ax5 = plt.subplots()

ax5.plot(flights_grouped["year"],flights_grouped["passengers"],
         color="yellow",ls="--",
         marker="o",markerfacecolor="orange",markeredgecolor="black")

ax5.set_xlabel("Year")
ax5.set_ylabel("Number of Passengers")
ax5.set_title("Total Passengers Per Year")

plt.show()
```

```python
# 5

# i)
tips = sns.load_dataset("tips")

print(tips.describe())
print(tips.info())
print(tips.head())
```

```python
# ii)
tips_grouped = tips.groupby("day")["tip"].mean().reset_index()

fig6, ax6 = plt.subplots()

ax6.bar(tips_grouped["day"],tips_grouped["tip"])

ax6.set_xlabel("Weekday")
ax6.set_ylabel("Average Tip")
ax6.set_title("Average Tip By Weekday")

plt.show()

# iii)
fig7, ax7 = plt.subplots()

ax7.hist(tips["total_bill"],bins=20,edgecolor="black")

ax7.set_xlabel("Total Bill")
ax7.set_ylabel("Frequency")
ax7.set_title("Distribution of Bill Sizes")

plt.show()
```
